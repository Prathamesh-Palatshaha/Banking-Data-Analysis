````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('/content/banking-clients.csv')
df.head()
````

````python
# Generate descriptive statistics
df.describe()
````

````python
df.info()
````

# Binning the estimated values into caterogorical values

````python
bins = [10000,100000, 150000,float('inf')]
labels = ['Low','Mid','High']
df['Income Band'] = pd.cut(df['Estimated Income'],bins=bins,labels=labels)
df['Income Band'].value_counts().plot(kind='bar')
````

````python
df.columns
````

# Analyzing catergorical Values or UNIVARIANT ANALYSIS - BECAUSE YOU ARE ANALYSING ONE COLUMN AT A TIME

````python
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns

for cols in categorical_cols:
  print(f"Value count of {cols}:{ display(df[cols].value_counts())}")
````

# BI-VARIANT ANALYSIS

````python
for col in categorical_cols:
    plt.figure()  # Create a new figure for each plot
    sns.countplot(data=df, x=col, hue='Nationality')
    plt.title(f'Count plot of {col} by Nationality') # Add a title for clarity
    plt.xticks(rotation=90, ha='right') # Rotate x-axis labels for better readability
    plt.tight_layout() # Adjust layout to prevent labels overlapping
    plt.show() # Display the figure
````

# NUMERICAL ANAYLSIS

````python
# Dropping the categorical columns
df_numerical = df.drop(columns=categorical_cols)

# Getting the numerical columns
numerical_cols = df_numerical.columns

# Calculating the number of rows needed for subplots
n_cols = 2
n_rows = (len(numerical_cols) + n_cols - 1) // n_cols

# Creating the subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, n_rows * 4))
axes = axes.flatten() # Flatten the axes array for easy iteration

# Plotting the numerical columns
for i, col in enumerate(numerical_cols):
    sns.histplot(data=df_numerical, x=col, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')

# Removing any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
````

````python
df_numerical = df.drop(columns=categorical_cols)
df_numerical = df_numerical.drop(columns=["Client ID",'Age','Location ID','Name','Joined Bank', 'Banking Contact'])

correlation_matrix = df_numerical.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title('Lower Triangular Correlation Heatmap')
plt.show()
````

## Insights
1. We can see 'Bank Deposits' has higher correlation with -->[Saving Accounts, Checking Accounts,Foreign Currency Account] which implies (=>) customer having higher bank balance is likely to have higher balance in other accounts too.
2. Higher balance indicates that bank is likely to lead money to those customers
